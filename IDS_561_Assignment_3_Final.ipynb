```python
#Creating  spark session
from pyspark.sql import SparkSession
spark = SparkSession.builder.master("local[*]").getOrCreate()

file = '/content/Amazon_new.csv'
df = spark.read.load(file,format="csv", inferSchema="true", header="true")
df.head(3)
```

```python
from pyspark import SparkContext
sc =SparkContext.getOrCreate()
```

```python
df_1 =df.select('tweet_created_at','user_screen_name','user_id_str')
print(df_1.count())

# Date format for further use
from pyspark.sql.functions import concat,col,isnan, when, trim,substring

df_2 = df_1.withColumn("tweet_created_at", concat(df_1.tweet_created_at.substr(4,8),df_1.tweet_created_at.substr(27,32)))
print(df_2.count())

#Removing duplicates
df_3=df_2.dropDuplicates()
print(df_3.count())

#Removing NA
df_4=df_3.dropna(subset=('tweet_created_at','user_id_str','user_screen_name'))
print(df_4.count())
```

```python
df_4.show()
```

```python
# Removing True & False values from dataset

df_4=df_4.filter(~df_4.user_screen_name.isin(['True','False']))
df_4=df_4.filter(~df_4.tweet_created_at.isin(['True','False']))
df_4=df_4.filter(~df_4.user_id_str.isin(['True','False']))
print(df_4.count())

#Keeping valid dates
df_4 = df_4.select("tweet_created_at", "user_screen_name", "User_id_str",
  substring(trim(col("tweet_created_at")),-4,4).cast("int").isNotNull().alias("Value"))
df_4 = df_4.where(df_4['Value'] == 'True')

#Removing User ID = 0
df_4 = df_4.where(df_4['User_id_str'] != 0)
print(df_4.count())

#Checking correct dates
df_4.orderBy("tweet_created_at", ascending=False).show()
```

Task 1: Create a dataframe “daily_active_users”. Find out the users who are active in at least five listed days (i.e., created posts in at least 5 days) in Amazon_Responded_Oct05.csv and and save their “user_screen_name” and “user_id_str” in the dataframe.

```python
#Step 1: Finding out users active for atleast 5 days

import pyspark.sql.functions as f

df_active_users = df_4.groupBy('user_screen_name').count().select('user_screen_name', f.col('count').alias('count_of_tweets'))
df_active_users=df_active_users.where(df_active_users.count_of_tweets > 5)
print(df_active_users.count())
df_active_users.show()
```

```python
user_id_screen_name = df_active_users.alias('a').join(df_4.alias('b'),col('a.user_screen_name') == col('b.user_screen_name'))
daily_active_users = user_id_screen_name.select(['a.user_screen_name', 'b.User_id_str']).dropDuplicates()

print(daily_active_users.count())                                              
daily_active_users.show()
```

Task 2: A company would like to conduct an A/B test on Twitter. The experiment.txt file includes the user_id_str they selected as potential experiment targets. Please create a dataframe “experiment_user” to document the select user id and whether they are active users (join the dataframe from step 1). Then calculate the percentage of active user and print out the result.

```python
#Step 2: A/B Testing on Twitter

file2 = '/content/11422218.txt'
expt = spark.read.load(file2,format="csv", inferSchema="true", header="false")

expt = expt.select(f.col('_c0').alias("experiment_User_id_str"))
print(expt.count())
expt.show()
```

```python
# Joining to find the active users

left_join = expt.join(daily_active_users, expt.experiment_User_id_str == daily_active_users.User_id_str,how='left')
Step2_Output = left_join.select('experiment_User_id_str','User_id_str')
print(Step2_Output.count()) 
Step2_Output.show()   
```

```python
from pyspark.sql import functions as f

sel_cols = ['experiment_User_id_str', f.when(f.col('User_id_str').isNull(),'Not_Active').otherwise('Active').alias('Active_Status')]

expt_user = Step2_Output.select(sel_cols)
print(expt_user.count())
expt_user.show()
```

% Active Users

```python
op1 = expt_user.where(expt_user['Active_Status'] == 'Active').count()/expt_user.count()*100
print(f'{op1}%')
```

```python
#Selecting all rows  for which user_id_str is present in  daily_active_user and expt_user

temp = df.alias('a').join(daily_active_users.alias('b'),col('a.User_id_str') == col('b.User_id_str'),how='inner')

op_r1= temp.select("a.*").dropDuplicates()
print(op_r1.count()) 
#op_r1.show()

op_s2 = expt_user.where(expt_user['Active_Status'] == 'Active')
print(op_s2.count()) 


inner_join = op_r1.alias('a').join(op_s2.alias('b'), col('a.User_id_str') == col('b.experiment_User_id_str'),how='inner')
final_op = inner_join.select("a.*").dropDuplicates()

print(final_op.count()) 
final_op.show()
```
